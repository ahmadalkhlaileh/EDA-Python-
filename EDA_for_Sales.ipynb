```python
def detect_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (df < lower_bound) | (df > upper_bound)

# Get numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns

# Detect outliers for each numerical column
for col in numerical_cols:
    outliers = detect_outliers_iqr(df[col])
    print("Number of outliers detected in column '" + col + "': " + str(sum(outliers)))
```

```python
numerical_cols = df.select_dtypes(include=np.number).columns


for col in numerical_cols:
    outliers = detect_outliers_iqr(df[col])
    if outliers.any():
        print("Outliers detected in column: " + col)
        print(df[col][outliers].tolist())
    else:
        print("No outliers detected in column: " + col)
```

```python

numeric_columns = df.select_dtypes(include=[np.number]).columns


num_plots = len(numeric_columns)
plt.figure(figsize=(15, num_plots * 3))


for i, col in enumerate(numeric_columns, 1):
    outliers = detect_outliers_iqr(df[col])


    plt.subplot(num_plots, 2, 2 * i - 1)
    plt.scatter(range(len(df)), df[col], c=['red' if x else 'blue' for x in outliers])
    plt.xlabel('Index')
    plt.ylabel(col)


    plt.subplot(num_plots, 2, 2 * i)
    sns.boxplot(x=df[col])
    plt.xlabel(col)


plt.tight_layout()
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Day"] = df["Date"].dt.day
df["Month"] = df["Date"].dt.month
df["Weekday"] = df["Date"].dt.day_name()


monthly_sales = df.groupby('Month')['Total'].sum()


plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```
